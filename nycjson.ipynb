```python
import json
import requests
from pymilvus import model
from pymilvus.model.dense import SentenceTransformerEmbeddingFunction

model = SentenceTransformerEmbeddingFunction('all-MiniLM-L6-v2',device='cpu' )

response = requests.get('https://data.cityofnewyork.us/resource/h9gi-nx95.json?$order=crash_date+DESC&$limit=10000').content

# for testing replace 5000 with 5
crashes = json.loads(response)
```

```python
import os
from pymilvus import MilvusClient

DIMENSION = 384 
MILVUS_URL = "http://192.168.1.163:19530" 
COLLECTION_NAME = "nyccollisions"

# -----------------------------------------------------------------------------
# Connect to Milvus

# Milvus Lite
# milvus_client = MilvusClient(uri="streetcams.db")

# Local Docker Server
milvus_client = MilvusClient( uri=MILVUS_URL )
```

```python
from pymilvus import connections
from pymilvus import utility
from pymilvus import FieldSchema, CollectionSchema, DataType, Collection
import pprint

## schema

schema = milvus_client.create_schema(
    auto_id=False,
    enable_dynamic_field=False
)

schema.add_field(field_name="id", datatype=DataType.INT64, is_primary=True)
schema.add_field(field_name="vector", datatype=DataType.FLOAT_VECTOR, dim=DIMENSION)
schema.add_field(field_name="crash", datatype=DataType.JSON)

index_params = milvus_client.prepare_index_params()

index_params.add_index(
    field_name="id",
    index_type="STL_SORT"
)

index_params.add_index(
    field_name="vector",
    index_type="IVF_FLAT",
    metric_type="L2",
    params={"nlist": 100}
)

milvus_client.create_collection(
    collection_name = COLLECTION_NAME,
    schema=schema,
    index_params=index_params
)

res = milvus_client.get_load_state(
    collection_name = COLLECTION_NAME
)

print(res)
```

```python
import numpy
fulladdress = ""
details = ""
location = ""
data = []
rowCount = 0

# print(model.dim)

for jsonitems in crashes:
     rowCount = rowCount + 1
     location = '{0} {1}, NY {2}'.format(jsonitems.get('on_street_name','Unknown Street'),  jsonitems.get('borough',''), jsonitems.get('zip_code','') )
     details = 'Crash occurred on {0} with off street {1} {2}, NY {3} with lat/long {4} {5} at {6} on {7} with vehicles {8} and {9} including {10} injuries'.format(
                                  jsonitems.get('on_street_name','Unknown Street'), jsonitems.get('off_street_name',''), jsonitems.get('borough',''),
                                  jsonitems.get('zip_code',''), jsonitems.get('latitude',''), jsonitems.get('longitude',''),
                                  jsonitems.get('crash_time',''),  jsonitems.get('crash_date',''), jsonitems.get('vehicle_type_code1',''), jsonitems.get('vehicle_type_code2',' '), jsonitems.get('number_of_persons_injured','0'))

     data.append({"id": numpy.int64(jsonitems.get('collision_id',rowCount)),
                  "vector": model(details),
                  "crash": {
                      "details": details,
                      "on_street_name": jsonitems.get('on_street_name','Unknown Street'),
                      "off_street_name": jsonitems.get('off_street_name',''),
                      "crash_date": jsonitems.get('crash_date',''),                                  
                      "crash_time": jsonitems.get('crash_time',''),                           
                      "borough": jsonitems.get('borough',''),                                      
                      "zip_code": jsonitems.get('zip_code',''),                                    
                      "latitude": jsonitems.get('latitude',''),                                    
                      "longitude": jsonitems.get('longitude',''),                                   
                      "location": location, 
                      "number_of_persons_injured": jsonitems.get('number_of_persons_injured',''),                     
                      "number_of_persons_killed": jsonitems.get('number_of_persons_killed',''),                      
                      "number_of_pedestrians_injured": jsonitems.get('number_of_pedestrians_injured',''),                 
                      "number_of_pedestrians_killed": jsonitems.get('number_of_pedestrians_killed',''),                  
                      "number_of_cyclist_injured": jsonitems.get('number_of_cyclist_injured',''),                     
                      "number_of_cyclist_killed": jsonitems.get('number_of_cyclist_killed',''),                      
                      "number_of_motorist_injured": jsonitems.get('number_of_motorist_injured',''),                    
                      "number_of_motorist_killed": jsonitems.get('number_of_motorist_killed',''),                     
                      "contributing_factor_vehicle_1": jsonitems.get('contributing_factor_vehicle_1',''),
                      "vehicle_type_code1": jsonitems.get('vehicle_type_code1',''),
                      "contributing_factor_vehicle_2": jsonitems.get('contributing_factor_vehicle_2',''),
                      "vehicle_type_code2": jsonitems.get('vehicle_type_code2',''),
                      "cross_street_name": jsonitems.get('cross_street_name',''),
                      "contributing_factor_vehicle_3": jsonitems.get('contributing_factor_vehicle_3',''),
                      "vehicle_type_code_3": jsonitems.get('vehicle_type_code_3',''),
                      "contributing_factor_vehicle_4": jsonitems.get('contributing_factor_vehicle_4',''),
                      "vehicle_type_code_4": jsonitems.get('vehicle_type_code_4','') 
                   } 
                 })

res = milvus_client.insert(collection_name=COLLECTION_NAME, data=data)
print(res)
```

```python
# Define search parameters
search_params = {
    "metric_type": "L2",
    "params": {"nprobe": 10}
}

# Use first record as search record
query_vector = [data[0]["vector"]]

# Execute the search on the 'vector' field
search_results = milvus_client.search(
    COLLECTION_NAME,
    data=query_vector,
    anns_field="vector",
    output_fields=["id", "crash"],
    search_params=search_params,
    limit=5
)

# Print search results
for hits in search_results:
    for hit in hits:
        print(f"Hit: {hit}")
```
